# Comparision of marks between Male and Female

```python
import matplotlib.pyplot as plt
import numpy as np
w=0.4
x=['Math_Score','Reading_Score','Writing_core']
FEMALE=[50,60,65.5]
MALE=[73,74,72]

bar1=np.arange(len(x))
bar2=[i+w for i in bar1]

plt.bar(bar1,FEMALE,w,label="FEMALE")
plt.bar(bar2,MALE,w,label="MALE")
plt.xlabel("Gender" )
plt.xlabel("Score")
plt.title("Group A Score")
plt.xticks(bar1,x)
plt.legend()
plt.show()


```
