```python
s=t2['CONSTITUENCY']
s1=list(s)
j=t2['TOTAL\nVOTES']
j1=list(j)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12,10))
plt.pie(j1,labels=s1,autopct='%1.2f%%')
plt.title('Votes cated to BJP in Delhi-NCR')
plt.show()
```
