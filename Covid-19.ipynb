# Month wise corona case prediction in year 2021

```python
import numpy as np

x=np.array([[1],[2],[3],[4]])

y=np.array([[5157],[11007],[225022],[555285]])
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x,y)
```

```python
coef=lr.coef_[0]
coef
```

```python
intercept=lr.intercept_
intercept
```

```python
exp=int(input('Type month number to know confirmed cases'))
pred_sal=coef*exp+intercept
pred_sal
```

```python
pred=lr.predict(x)
pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,pred)
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,pred))
```

```python
import numpy
x=np.array([1,2,3,4])
y=np.array([5157,11007,225022,555285])

mymodel=numpy.poly1d(numpy.polyfit(x,y,5))

pred=mymodel(x)
pred

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,pred)
plt.show()

from sklearn.metrics import r2_score
print(r2_score(y,pred))
```
