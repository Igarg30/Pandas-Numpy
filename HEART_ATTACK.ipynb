```python
import matplotlib.pyplot as plt
import numpy as np
w=0.4
x=['ASY','ATA','NAP','TA']
FEMALE=[14.11,34.68,26.11,21.74]
MALE=[85.89,65.32,73.89,78.26]

bar1=np.arange(len(x))
bar2=[i+w for i in bar1]

plt.bar(bar1,FEMALE,w,label="FEMALE")
plt.bar(bar2,MALE,w,label="MALE")
plt.xlabel("Chest Pain")
plt.xlabel("Sex")
plt.title("Chest Pain in Male & Female")
plt.xticks(bar1,x)
plt.legend()
plt.show()

```

# HIGHEST RESTING BP IN ALL CHEST PAIN TYPE

```python
s=is3['ChestPainType']
s1=list(s)
j=is3['RestingBP']
j1=list(j)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.bar(s1,j1,color='red')
plt.xlabel('Chest Pain')
plt.ylabel('Resting BP')
plt.title('Highest Resting BP in all Chest pain type')
plt.show()

```
