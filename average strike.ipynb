```python
import pandas as pd
df=pd.read_excel("C:/Users/Admin/Desktop/ipl data/average strike - Copy.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.fillna({'average':0},inplace=True)
```

```python
df.isnull().sum()
```

```python
df1=df[df['strikerate']>170]
```

```python
#Batsman whose strike rate are above 170
s=df1['batsman']
s1=list(s)
j=df1['strikerate']
j1=list(j)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12,10))
plt.pie(j1,labels=s1,autopct='%1.2f%%')
plt.title('Batsman whose strike rate are above 170')
plt.show()
```

```python
df2=df[df['average']>30]
df3=df2[df['out']>100]
df3
```

```python
#batsman whose average is more than 30 and out is more than 100
s=df3['average']
s1=list(s)
j=df3['batsman']
j1=list(j)
fig=plt.figure(figsize=(12,5))
plt.bar(j1,s1,color='red')
plt.xlabel('PLAYER NAME')
plt.ylabel('FREQUENCY')
plt.title('Player whose average is more than 30 and out is more than 100')

plt.show()
```

```python
df4=df[df['total_runs']>4300]
df5=df4[df['out']>100]
df5
```

```python
# scatter graph for players who score more than 4300 and out is more than 100
s=df5['total_runs']
s1=list(s)
j=df5['out']
j1=list(j)
plt.scatter(j1,s1,color='red',marker='v')
plt.xlabel('No.of out')
plt.ylabel('Total Run')
plt.title('players who score more than 4300 and out is more than 100')
plt.show()
```
